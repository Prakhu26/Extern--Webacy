```python
data= df.fillna(0)
```

```python
# List of columns representing the risk tags
risk_columns = ['Is_closed_source', 'hidden_owner', 'anti_whale_modifiable', 'Is_anti_whale', 
                'Is_honeypot', 'buy_tax', 'sell_tax', 'slippage_modifiable', 'Is_blacklisted',
                'can_take_back_ownership', 'owner_change_balance', 'is_airdrop_scam', 'selfdestruct',
                'trust_list', 'is_whitelisted', 'is_fake_token', 'illegal_unicode', 'exploitation',
                'bad_contract', 'reusing_state_variable', 'encode_packed_collision',
                'encode_packed_parameters', 'centralized_risk_medium', 'centralized_risk_high',
                'centralized_risk_low', 'event_setter', 'external_dependencies', 'immutable_states',
                'reentrancy_without_eth_transfer', 'incorrect_inheritance_order', 'shadowing_local', 'events_maths']
```

```python
# Ensure the columns are converted to boolean type
data[risk_columns] = data[risk_columns].astype(bool)
```

```python
# Frequency Distribution of True values in each risk tag
frequencies = data[risk_columns].apply(lambda x: x.value_counts()).loc[True]
```

```python
# Plotting frequency distribution
plt.figure(figsize=(12, 8))
sns.barplot(x=frequencies.index, y=frequencies.values)
plt.title('Frequency of True Values for Each Risk Tag')
plt.xlabel('Risk Tags')
plt.ylabel('Frequency of True')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Phi Coefficient Calculation for Binary Risk Tags
def phi_coefficient(x, y):
    """Calculate the Phi coefficient for two binary variables."""
    # Create a contingency table
    contingency_table = pd.crosstab(x, y)
     # Calculate the phi coefficient from the contingency table
    chi2 = scipy.stats.chi2_contingency(contingency_table, correction=False)[0]
    # Sum the contingency table to get the total number of observations
    n = contingency_table.sum().sum()  # Sum over both axes
    # Calculate the phi coefficient
    phi = np.sqrt(chi2 / n)
    return phi
```

```python
# Calculate Phi coefficient for each pair of binary variables and store in a matrix
phi_matrix = pd.DataFrame(index=risk_columns, columns=risk_columns)
for var1 in risk_columns:
    for var2 in risk_columns:
        phi_matrix.loc[var1, var2] = phi_coefficient(df[var1], df[var2])
# Convert to float for correct heatmap display
phi_matrix = phi_matrix.astype(float)
```

```python
# Plot the heatmap of Phi Coefficient Correlations
plt.figure(figsize=(16, 10))
sns.heatmap(phi_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Phi Coefficient Heatmap of Risk Tags')
plt.show()
```
